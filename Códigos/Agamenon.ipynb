## 1.2 Tratando os arquivos

Arquivo usado não está codado em utf-8 e por isso ele não lê o arquivos, precisamos resolver esse problema:

Vamos tentar decodificá-los diretamente, e em seguida tentar decodificar com uma abordagem mais personalizada:

```python
# Decodificar o arquivo sem descompactá-lo
zip_path = r"C:\Users\henri\Documents\GitHub\DataScience2\Arquivos\archive.zip"

with zipfile.ZipFile(zip_path, 'r') as z:
    for info in z.infolist():
        try:
            # Tentando decodificar o nome do arquivo
            print(info.filename.encode('cp437').decode('utf-8')) # '.filename' retorna o nome do arquivo ZIP
        except UnicodeDecodeError:
            # Se der erro de decodificação, imprimimos o nome como está
            print(f"Erro ao decodificar o nome do arquivo: {info.filename}")
```

Conseguimos decodificar o arquivo, acima

```python
zip_path = r"C:\Users\henri\Documents\GitHub\DataScience2\Arquivos\archive.zip"

with zipfile.ZipFile(zip_path, 'r') as z:
    with z.open('dataset-2003-2019.csv') as p:
        df = pd.read_csv(p, sep=';')
```
